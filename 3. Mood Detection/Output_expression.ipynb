```python
import numpy as np
import dlib
import cv2
from sklearn.neighbors import KNeighborsClassifier

detector = dlib.get_frontal_face_detector()

predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")



data = np.load("face_data.npy")

X = data[:, 1: ].astype(int)   #all rows and all columns from the first one
y = data[:, 0]    #all rows and 0th column.
#print(X.type)

model = KNeighborsClassifier()
model.fit(X, y)

cap = cv2.VideoCapture(0) 

while True:
    
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = detector(gray)

    for face in faces:
        landmarks = predictor(gray, face)
        
        #to get the predictions for the features we flatten the features and we want the points to start from the initial starting ie top left corner , so we modified values by sutracting.
        expression = np.array([[point.x - face.left(), point.y - face.top()] for point in landmarks.parts()[17:]])
        
        print(model.predict([expression.flatten()]))
        
        #out = model.predict([expression.flatten()])
        #cv2.putText(frame, out, (20, 20), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 255), 1)
        
    
    if ret:
        cv2.imshow("My Screen", frame)

    key = cv2.waitKey(1)
    if key == ord("q"):
        break
    

cap.release()
cv2.destroyAllWindows()
```
