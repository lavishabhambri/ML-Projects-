# Example of classification dataset for visualisation

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 100, centers = 3, random_state = 42)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c = y)    #X[:, 0], X[:, 1] are the features x1 and x2 and the classes are represented on the basis of colors.
```
